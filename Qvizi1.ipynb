```python
data.set_index("ResponseId", inplace=True)
```

```python
მოცემულ ეტაპზე ხდება RsponseId ითვის ინდექსის მინიჭება რაც საშუალებას მაძლევს ჩანაწერები ResponseId მიხედვით უფრო სწრაფად მოვძებნო და არ ვიყო დამოკიდებული default ინდექსზე
inplace=true საშუალებას მაძლევს რომ ჩემი dataFrame ახალ ცვლადში არ შევინახო და იმ frame ზე არ ვიმუშაო და არსებულ ცხრილს უკეთებს ინდექსირებას წინააღმდეგ შემთხვევაში 
დამჭირდებოდა ცვლადის შემოღება
data.reset_index(inplace=True) ეს კი საშუალებას მომცემს დავაბრუნო default ინდექსირებაზე იგივე data frame ში
```

```python
filtered_data = data[(data['Age'] >= '25-34 years old') & (data['Employment'] == 'Employed, full-time')]
print(filtered_data)
```

```python
მოცემულ ეტაპზე ხდება age მიხედვით ფილტრაცია სადაც ასაკს ვამოწმებ არის თუ არა მეტი 25-34 აღნიშნულ მონაცემებს ვადებ ფილტრს Employment მიხედვით სადაც ვირებ მხოლოდ იმათ ვინც
აკმაყოფილებს პირობას რომ მუშოაბს Employed, full-time აღებლ ინფორმაციას ვსეტავ filtered_data ცვლადში და გამომაქვს ვიზუალურად
```

```python
filter1 = (data['Age'] == '18-24 years old') & (data['EdLevel'] == "Bachelor’s degree (B.A., B.S., B.Eng., etc.)")
filter2 = (data['Age'] == '25-34 years old') & (data['EdLevel'] == "Master’s degree (M.A., M.S., M.Eng., etc.)")
filtered_data = data[filter1 | filter2]

print(filtered_data)
```

```python
ამ ეტაპზე როგორც წინა სვეტშ იყო ვიღებ ორი ფილტრის მიხდევით მონაცემებს განსხვავება ისაა რომ ამ ორ ფილტრს or ფუნქციით ვაერთიანებ და ისე გამოვსახავ შედეგს
```

```python
sorted_data = data.sort_values(by=['Age', 'EdLevel'], ascending=[True, False])
print(sorted_data)
```

```python
ამ ეტაპზე ვახდენ მონაცემების დალაგებას სადაც Age არის დალაგებული ზრდადობით ხოლო EdLevel კლებადობით(ალფაბითის მიხედვით ალაგებს ამ შემთხვევაში)
```

```python
data['ConvertedCompYearly'] = pd.to_numeric(data['ConvertedCompYearly'], errors='coerce')

filtered_data = data.dropna(subset=['ConvertedCompYearly'])

comp_yearly_mean = filtered_data['ConvertedCompYearly'].mean()
comp_yearly_std = filtered_data['ConvertedCompYearly'].std()
comp_yearly_median = filtered_data['ConvertedCompYearly'].median()
comp_yearly_min = filtered_data['ConvertedCompYearly'].min()
comp_yearly_max = filtered_data['ConvertedCompYearly'].max()
comp_yearly_max_formatted = "{:,.0f}".format(comp_yearly_max)

print("ConvertedCompYearly Max:", comp_yearly_max_formatted)
print("ConvertedCompYearly Mean:", comp_yearly_mean)
print("ConvertedCompYearly Standard Deviation:", comp_yearly_std)
print("ConvertedCompYearly Median:", comp_yearly_median)
print("ConvertedCompYearly Min:", comp_yearly_min)
```

```python
მოცემულ კოდში ხდება თავდაპირველად შემოწმება თუ ხვდება ისეთი მნიშვენლობები რომელიც არ არის რიცხვითი რომ პრობლემა არ გამოიწიოს შემდეგში, მომდევნო ნაწილში წინა ეტაპზე დახარისხებულ მონაცემებს იყენებს და ამოწმებს
შეიცავს თუ არა აღნიშნული NAN მნიშვენლობებს რომ მოაშოროს ისინი მომდევნო ეტაპებზე ხდება დამუშაოებული ინფრომაციებთან კონკრეტული ფუნქციების გამოყენება და ამ ყველაფრის ჩვენება ვიზუალურად
```

```python
age_counts = data['Age'].value_counts().sort_index()

plt.figure(figsize=(10, 6))
age_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Age')
plt.ylabel('Number of Respondents')
plt.title('Age Distribution of Respondents')
plt.show()
```

```python
data['ConvertedCompYearly'] = pd.to_numeric(data['ConvertedCompYearly'], errors='coerce')

data.dropna(subset=['ConvertedCompYearly'], inplace=True)

plt.figure(figsize=(10, 6))
plt.plot(data['ConvertedCompYearly'], color='green')
plt.xlabel('Respondent Index')
plt.ylabel('Yearly Compensation (USD)')
plt.title('Yearly Compensation of Respondents')
plt.show()
```
